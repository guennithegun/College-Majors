### scatter plots
now exploring some relations

```python
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', title='Employed vs. Sample_size', figsize=(16,9), xlim=(0,5000), ylim=(0,350000))
```

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter', figsize=(16,9))
ax.set_title('Median vs. Sample_size')
ax.set_xlim(0,5000)
```

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', figsize=(16,9))
ax.set_title('Unemployment Rate vs. Sample_size')
ax.set_xlim(0,5000)
ax.set_ylim(0, 0.2)
```

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter', figsize=(16,9))
ax.set_title('Full_time vs. Median')
ax.set_xlim(0,300000)
```

```python
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', figsize=(16,9))
ax.set_title('ShareWoman vs. Unemployment_rate')
ax.set_xlim(0, 1.2)
ax.set_ylim(0, 0.2)
```

```python
ax = recent_grads.plot(x='Men', y='Median', kind='scatter', figsize=(16,9))
ax.set_title('Men vs. Median')
ax.set_xlim(0,200000)
```

```python
ax = recent_grads.plot(x='Women', y='Median', kind='scatter', figsize=(16,9))
ax.set_title('Women vs. Median')
ax.set_xlim(0,350000)
```
